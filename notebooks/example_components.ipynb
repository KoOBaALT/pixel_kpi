# Component Library

This notebook is a collection of pixel-art UI-components that can be used to create a pixelart KPI dashboards. The following components are included:
- **Text**: A component to display text
- **Headline**: A component to display a headline
- **Bar Chart**: A component to display a bar chart
- **Progress Bar**: A component to display a progress bar
- **Star Icon**: A component to display a star icon
- **Image**: A component to display an image
- **Fireworks with Text**: A component to display fireworks with text

Those components can be used to create a pixel art view, which could be displayed on a Pixoo or on a normal screen. For more information of how to build custom views have a look how the view components are build (pixel_kpi/views/)

## Initialization of a Display

The package `pixel_kpi` provides a class `Display` that can be used to display the components. THose displays could be for example a Pixoo or a normal screen. The following code shows how to initialize a display:

```python
from pixel_kpi.displays.matplotlib_display import MatplotlibDisplay
from pixel_kpi.displays.pixoo_display import PixooDisplay

display = MatplotlibDisplay(display_name="MatplotlibDisplay", 
                                    display_id=0, 
                                    width=64, 
                                    height=64,
                                    background_color=(0, 0, 0))
#IP_ADDRESS = "192.168.178.153"
#
#display = PixooDisplay(display_name="PixooDisplay",
#                        display_id=0,
#                        width=64,
#                        height=64,
#                        background_color=(0, 0, 0),
#                        ip_address=IP_ADDRESS,
#                        simulated=False)

component_main_color = (0, 255, 255)
```

## Text Component

```python
from pixel_kpi.components.text import Text

display.clear()

Text.draw(display, text="Text", position=(24, 30), color=component_main_color)

display.render()
```

## Headline Component

```python
from pixel_kpi.components.headline import Headline

display.clear()

Headline.draw(display, text="Headline", position=(1, 25), scale=2, color=component_main_color)

display.render()
```

## Bar Chart Component

```python
from pixel_kpi.components.bar_chart import BarChart

display.clear()

BarChart.draw(display, 
              data=[2, 2, 3, 4, 5, 6, 7, 8, 9, 9, 21, 30], 
              labels=["J", "F", "M", "A", "M", "J", "J", "A", "S", "O", "N", "D"], 
              position=(3, 15),
              max_height=20, 
              bar_width=3,
              spacing=2,
              color=component_main_color)

display.render()
```

## Progress Bar Component

```python
from pixel_kpi.components.progress_bar import ProgressBar

display.clear()

ProgressBar.draw(display, 
              title="Progress", 
              progress=0.2, 
              position=(3, 24),
              color=component_main_color)

display.render()
```

## Star Icon Component

```python
from pixel_kpi.components.star_icon import StarIcon

display.clear()

StarIcon.draw(display, position=(3, 3), color=component_main_color)

display.render()
```

## Image Component

```python
from pixel_kpi.components.fullscreen_image import FullscreenImage

display.clear()

FullscreenImage.draw(display, path="assets/example_img.png")

display.render()
```

## Fireworks with Text Component

```python
from pixel_kpi.components.firework import Firework
from pixel_kpi.components.text import Text
from IPython.display import clear_output

import time

display.clear()

firework_1 = Firework(display=display, max_height=50, position=(64, 32))
firework_2 = Firework(display=display, max_height=30, position=(64, 15))
firework_3 = Firework(display=display, max_height=47, position=(64, 20))
firework_4 = Firework(display=display, max_height=50, position=(64, 23))
firework_5 = Firework(display=display, max_height=53, position=(64, 56))

for j in range(3):
    for i in range(200):
        
        Text.draw(display, text="Sprint 1", position=(18, 20), color=(255, 255, 255))
        Text.draw(display, text="Completed!", position=(14, 30), color=(255, 255, 255))

        if i > 2:
            firework_1.draw()
        if i > 40:
            firework_2.draw()
        if i > 60:
            firework_3.draw()
        if i > 90:
            firework_4.draw()
        if i > 100:
            firework_5.draw()
        time.sleep(0.05)
        display.render()
        clear_output(wait=True)
```
